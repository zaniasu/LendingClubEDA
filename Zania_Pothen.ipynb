Problem Statement - Identify key variables that can be used by the consumer finance company can use to identify applicants that are likely to default on a loan.


```python
# Loading the data given
import pandas as pd
cust_details = pd.read_csv('/Users/zsp/Downloads/loan.csv',low_memory=False)
cust_details.info()
cust_details.head(3)
```

The inital dataset contains 111 columns and a total of 39717 row entries. First we identify if all those 111 columns contain usefull information - if they do not then we remove them from further consideration.

```python
print(cust_details.isnull().sum())
```

```python
# remove columns where no data is provided
cust_details.dropna(how='all', axis=1, inplace=True)
```

```python
cust_details.info()
cust_details.head(3)
```

```python
# Remove columns with constant values as these will not provide usefull information for the analysis
# # cust_details.to_csv('loan_modified.csv')
# print(cust_details.columns[cust_details.nunique() <= 1])
# # print(cust_details.loc[:, list(cust_details.columns[cust_details.nunique() <= 1])])
cust_details_mod=cust_details.drop(cust_details.loc[:, list(cust_details.columns[cust_details.nunique() <= 1])], axis=1)
```

```python
cust_details_mod.info()
cust_details_mod.head(3)
```

While examing the rows I found that rows where the 'total_pymt' was zero more that 10 other columns in those rows also contained zeros. Which seemed odd since this data seemed to extracted on 2016 for 2008-2011 data and unless there was a loss of data there should have been some entry in these columns.

```python
#remove rows where the relevant data is missing
# cust_details_mod['total_pymnt_inv'].isnull()
cust_details_mod.loc[cust_details_mod['total_pymnt'] == 0]
```

```python
cust_details_mod.drop(cust_details_mod.loc[cust_details_mod['total_pymnt']==0].index, inplace=True)
```

```python
# cust_details_mod.to_csv('loan_modified_2.csv')
```

```python
cust_details_mod.info()
```

On further consideration I decided to remove the rows where the loan_status is 'current' as these are still active it gives us little information on whether or not these customers would default or not.

```python
cust_details_mod.drop(cust_details_mod.loc[cust_details_mod['loan_status']=='Current'].index, inplace=True)
```

```python
cust_details_mod.info()
```

After the data cleaning steps I was left with the above 48 columns and 38561 rows.
Next I segment the data based on 'loan status'

```python
df_pv=pd.pivot_table(data=cust_details_mod,index='loan_status',aggfunc='size')
print((df_pv/df_pv.sum())*100)
```

From the above figure we can see that 85% repaid their loans but the remaining 16% defaulted.
Next, I will go through the list of variabale and try and figure out which ones can be used to predict defaulters.
First we examin the type of home ownership

```python
pd.crosstab(cust_details_mod.loan_status,cust_details_mod.home_ownership,margins=True,normalize='index')
```

Comparing the percentage of Mortage and renters in both 'Charged off' & 'Fully Paid' categories we can see
that the majority of borowers fall into these 2 categories and therefore cannot be used identify defaulters

```python
pd.crosstab(cust_details_mod.loan_status,cust_details_mod.verification_status,margins=True,normalize='index')
```

Verification also does not seem to be an indicator

```python
pd.crosstab(cust_details_mod.purpose,cust_details_mod.loan_status,margins=True,normalize='columns')
```

From the above data we can see that majority of borrowers are take a loan for debt_consolidation. 
Majority of defaulters fall into this category ~50%. So this means that these applicants should be scrutinized a bit more.

```python
pd.crosstab(cust_details_mod.purpose,cust_details_mod.loan_status,margins=True,normalize='index')
```
